```python
auto.rename(columns = {"id_cliente": "id_client"}, inplace=True)
```

```python
join = pd.merge(clientes, auto, on='id_client')
```

```python
join["city"] = join["city"].str.lower()
```

```python
total_cidade_ford = []

for city in join['city'].unique():
  total_cidade_ford.append('quantidade: ' + str(len(join.loc[(join['auto_brand'] == 'Ford') & (join['city'] == city)])) + ' cidade: ' + str(city))
```

```python
for s in total_cidade_ford:
  print(s)
```
